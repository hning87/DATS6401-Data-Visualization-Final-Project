```python
import numpy as np
import pandas as pd
import pycountry
```

```python
patent_r = pd.read_excel('Patent applications, residents.xls',header = 3,index_col = 0)
patent_nr = pd.read_excel('Patent applications, nonresidents.xls',header = 3,index_col = 0)
manufac =  pd.read_excel('Manufacturing, value added (current US$).xls',header = 3,index_col = 0)
publication = pd.read_excel('Scientific and technical journal articles.xls',header = 3,index_col = 0)
r_expense_percent = pd.read_excel('Research and development expenditure (% of GDP).xls',header = 3,index_col = 0)
gdp = pd.read_excel('GDP (current US$).xls',header = 3, index_col = 0)
```

```python
# get country_code
country_name, country_code = [],[]
for country in pycountry.countries: 
    country_name.append(country.name)
    country_code.append(country.alpha_3)
    
len(country_name)
```

```python
# set index as country code, drop region and area (that are not in country_code), leaving only countries in dataframe
publication.new= publication.reset_index().set_index('Country Code')

# creat a droplist 
droplist = []
for x in publication.new.index:
    if x not in country_code:
        droplist.append(x)
        
len(droplist)
```

```python
def cleaning(df):
    df_new = df.reset_index().set_index('Country Code')
    df_new = df_new.drop(droplist)
    df_new = df_new.reset_index().set_index('Country Name').rename(columns={'Country Code':'Code'})
    return df_new
    
# patent.new = patent.new.drop(droplist)

# # rename, reset index 
# patent.new = patent.new.rename(columns={'Country Name':'Country'}).reset_index().set_index('Country').rename(columns={'Country Code':'Code'})
# patent.new.head()
```

```python
patent_r = cleaning(patent_r)
patent_nr = cleaning(patent_nr)
manufac =  cleaning(manufac)
publication = cleaning(publication)
r_expense_percent = cleaning(r_expense_percent)
gdp = cleaning(gdp)
```

```python
patent = patent_r + patent_nr
```

```python
# loc countries in dataframe, drop region and area
df_patent = patent.loc[:, '2011':'2017'].dropna()
df_manufac =  manufac.loc[:, '2011':'2017'].dropna()
df_publication = publication.loc[:, '2011':'2017'].dropna()
df_r_expense_percent = r_expense_percent.loc[:, '2011':'2017'].dropna()
df_gdp = gdp.loc[:, '2011':'2017'].dropna()
df_r_expense = (df_r_expense_percent* df_gdp).dropna()
```

```python
df_r_expense_percent.head()
```

```python
# absolute change
patent_change = df_patent.diff(axis=1).fillna(value=0)
manufac_change =  df_manufac.diff(axis=1).fillna(value=0)
publication_change = df_publication.diff(axis=1).fillna(value=0)
r_expense_percent_change = df_r_expense_percent.diff(axis=1).fillna(value=0)
r_expense_change = df_r_expense.diff(axis=1).fillna(value=0)
```

```python
# percentage change
patent_change_p = (df_patent.pct_change(axis=1) * 100).fillna(value=0)
manufac_change_p =  (df_manufac.pct_change(axis=1) * 100).fillna(value=0)
publication_change_p = (df_publication.pct_change(axis=1) * 100).fillna(value=0)
r_expense_percent_change_p = (df_r_expense_percent.pct_change(axis=1) * 100).fillna(value=0)
r_expense_change_p = (df_r_expense.pct_change(axis=1) * 100).fillna(value=0)
```

```python
df_patent['sum'] = df_patent.sum(axis=1)
df_manufac['sum'] =  df_manufac.sum(axis=1)
df_publication['sum'] = df_publication.sum(axis=1)
df_r_expense_percent['sum'] = df_r_expense_percent.sum(axis=1)
df_gdp['sum'] = df_gdp.sum(axis=1)
df_r_expense['sum'] = df_r_expense.sum(axis=1)
```

```python
df_patent.sort_values('sum', inplace=True, ascending=False)
df_manufac.sort_values('sum', inplace=True, ascending=False)
df_publication.sort_values('sum', inplace=True, ascending=False)
df_r_expense_percent.sort_values('sum', inplace=True, ascending=False)
df_gdp.sort_values('sum', inplace=True, ascending=False)
df_r_expense.sort_values('sum', inplace=True, ascending=False)
```

```python
# reshape
def reshape_data(df):
    #use stack to change the data frame shape
    df_new = df.stack()
    df_new = pd.DataFrame(df_new)
    df_new = df_new.reset_index()
    #rename column names
    df_new = df_new.rename(columns = {"level_1":"Year",0:"Amount"})
    return df_new
```

```python
def get_top_countries(df, topnumber):
    df_new = df[0: topnumber].drop(columns='sum')
    return df_new
```

```python
# def fucntion for country code
def add_code(df):
    code = []
    for name in df['Country Name']:
        for i in range(len(country_name)):
            if name == country_name[i]:
                code.append(country_code[i])
    df['code'] = code
    return df
```
